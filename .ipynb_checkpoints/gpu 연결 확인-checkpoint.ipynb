9. tensorflow 설치 
    * "pip install tensorflow"은 tensorflow 버전 안맞게 설치됨
    * 따라서 [버전 확인](https://www.tensorflow.org/install/source_windows#tested_build_configurations) 필요
    * 가장 최신인 2.7.0 설치해봄
    > pip install tensorflow==2.7.0
    
    
    
9. 아래 수행해서 gpu 잡히는지 확인했더니 됨
<br><br>
10. pytorch 다운
    * [CUDA에 맞는 pytorch 다운](https://pytorch.org/get-started/previous-versions/)
    * cuda 버전이 11.7이어도 이거보다 낮기만 하면 됨<br>
    따라서 지금 최신 버전으로 다운
    
> conda install pytorch==1.10.1 torchvision==0.11.2 torchaudio==0.10.1 cudatoolkit=11.3 -c pytorch -c conda-forge
<br><br>
11. pytorch 에러
    * 만약 설치했는데 잘 안된다면 여러 pytorch를 다운받아서 충돌나서 안될 가능성이 있음. 
    * 따라서 가상환경 패키지 폴더로 가서 torch 관련 폴더를 몇개 지워주기
    * 현재 가상환경(py37) 기준 폴더 이동
    > C:\Users\MNI\anaconda3\envs\py37\Lib\site-packages
    
    * torch 파일 빼고는 비슷하게 생긴 파일 다 지우고 다시 설치함
